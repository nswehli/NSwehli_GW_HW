```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file="Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
df= pd.read_csv(file)
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
df.columns
df_reduced=df.iloc[:, [0, 1, 2, 3, 4, 6,7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
df_reduced.columns
```

```python
# Change reading 0 and 1 to No and Yes, respectively
df_reduced=df_reduced.replace({0: "No", 1:"Yes"})
df_reduced.head()
```

```python
# Extract rows for only those who attended a bootcamp
BTatended = df_reduced.loc[(df_reduced["AttendedBootcamp"] == "Yes")]

BTatended.head()

```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
BTatended=BTatended.replace('Free Code Camp is not a bootcamp - please scroll up and change answer to "no"','Free Code Camp')
camps =BTatended.groupby(["BootcampName"])
camps=camps.count()
camps=camps.sort_values("AttendedBootcamp",ascending=False)
camps=camps.loc[(camps["AttendedBootcamp"] > 10)]

camps=camps["AttendedBootcamp"]

camps.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
BTrecommend = BTatended.loc[(BTatended["BootcampRecommend"] == "Yes")]
recom =BTrecommend.groupby(["BootcampName"])
recom=recom.count()
recom=recom.sort_values("BootcampRecommend",ascending=False)
#recom=recom.loc[(recom["BootcampRecommend"] > 10)]
recom=recom ["BootcampRecommend"]
recom.head()
```

```python
# Merge the two created data frames on the name of tbe bootcamp
merge_table = pd.merge(camps, recom, on="BootcampName")

merge_table.head()
```

```python
# Calculate percentage of eac bootcamp's students who are recommenders
merge_table["Percentage of recommendations"]=((merge_table["BootcampRecommend"]/merge_table["AttendedBootcamp"])*100)
merge_table
```

```python
# Export to excel and remove index

merge_table=merge_table.sort_values("Percentage of recommendations",ascending=False)
report=merge_table.reset_index()
report["Percentage of recommendations"]=report["Percentage of recommendations"].map("${:.2f}%".format)

report.head()
```

```python

# Export to Excel
report.to_csv("output\BootcampOutputPart2.csv")
report.head()
```
