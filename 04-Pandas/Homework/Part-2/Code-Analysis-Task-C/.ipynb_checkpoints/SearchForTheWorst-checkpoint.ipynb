**Example**: The cell below creates a reference to teh csv file and stores it in a variable called `csv_path`. It then creates a DataFrame using the `pd.read_csv` function. The output of the `read_csv` function is a DataFrame.

```python
# Create reference to CSV file
csv_path = "Resources/Soccer2018Data.csv"

# Import the CSV into a pandas DataFrame
soccer_2018_df = pd.read_csv(csv_path, low_memory=False)
soccer_2018_df.head(3)
```

the .unique() function will return all the values that are listed in the ["Preferred Position"] which will return the prefered field position of the players. There will be no duplicates in the returned list. 

```python
soccer_2018_df["Preferred Position"].unique()
```

This cell use the .loc function to set a condition that only the rows that has "ST" listed in the "Preferred Position" column. 

```python
strikers_2018_df = soccer_2018_df.loc[soccer_2018_df["Preferred Position"] == "ST", :]
strikers_2018_df.head(2)
```

Using the sort_values function applied to the "ST" column, the "strikers_2018_df" will be begin with the player with the least "ST" score. Then the index reset will zero the index so it starts in the ranking order.

```python
# Sort the DataFrame by the values in the "ST" column to find the worst
strikers_2018_df = strikers_2018_df.sort_values("ST")

# Reset the index so that the index is now based on the sorting locations
strikers_2018_df = strikers_2018_df.reset_index(drop=True)

strikers_2018_df.head(2)
```

This cell use the .loc function to specify that only row index 0 will be returned. Applied to the strikers_2018_df , this will return the player with the least 'ST" score.

```python
worst_striker = strikers_2018_df.loc[0, :]
worst_striker
```
