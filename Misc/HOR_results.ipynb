```python
from splinter import Browser
from bs4 import BeautifulSoup as BS
import requests
from splinter import Browser
import re
import pandas as pd
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
for y in range(1,14):
    link = f"http://results.hnec.ly/cor/constituency/{y}"
    browser.visit(link)
    links_found = browser.links.find_by_partial_href('ballot')
    for x in links_found:
        url=(x['href'])
        print(url)
        city=(x.value)
        if city != "عام" and city != "نساء":
            
            try:
                table=pd.read_html(url)
                results=table[1]
                results["constituency"]=y
                results["city"]=city
                results["link"]=url
                results=results.drop_duplicates()
                results.to_csv('Results.csv', mode='a', index=False, header=False, encoding='utf-8-sig')
            except:
                print("error:", {url})
                continue
```
