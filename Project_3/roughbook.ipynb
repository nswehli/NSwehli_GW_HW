```python
import pandas as pd
```

```python
file="data\Scoring.csv"
scoringData=pd.read_csv(file)
```

```python
scoringData.sort_values(by=['year'],ascending=False).head(5)
```

```python
scoringData.sort_values(by='G',ascending=False).head(5)
```

```python
allTime=scoringData.groupby("playerID").sum()
allTime.head(5)
```

```python
X = allTime[["GP", "A", "Pts"]]
y = allTime["G"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
score = model.score(X, y)

est = model.fit(X, y)


print(est.summary())
```
