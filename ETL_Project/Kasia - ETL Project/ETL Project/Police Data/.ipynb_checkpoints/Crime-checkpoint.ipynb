```python
# Dependencies
import pandas as pd
import numpy as np
import datetime
```

```python
# File to Load (Bethesda location)

bethesda_df = "Resources/Crime_Bethesda.csv"

# Read the modified GoodReads csv and store into Pandas DataFrame (Bethesda location)

bethesda_df = pd.read_csv(bethesda_df, encoding="utf-8", low_memory=False)
bethesda_df.head()
```

```python
# Create new data with select columns (Bethesda location)

new_bethesda_df = bethesda_df[['Crime Name2', 'Start_Date_Time', 'Latitude', 'Longitude']].copy()
new_bethesda_df.head()
```

```python
# Create new column with a year only (Bethesda location)

new_bethesda_df['Year'] = pd.DatetimeIndex(new_bethesda_df['Start_Date_Time']).year
new_bethesda_df.head()
```

```python
# Copy relevant columns (Bethesda location)

new2_bethesda_df = new_bethesda_df[['Crime Name2', 'Year', 'Latitude', 'Longitude']].copy()

# Rename the columns (Bethesda location)

new2_bethesda_df.rename(columns = {"Crime Name2": "Crime Type", "Latitude": "Latitude (Bethesda)", "Longitude": "Longitude (Bethesda)"}, inplace = True)
```

```python
# FINAL OUTPUT (BETHESDA LOCATION)

new2_bethesda_df.head()
```

```python
# Output the csv file, so it can be uploaded to the database

new2_bethesda_df.to_csv('Modified_Crime_Bethesda.csv')
```
