```python
import pandas as pd
```

```python
file="data/Master.csv"
data=pd.read_csv(file)
names=data.groupby(["lastName"])
namesList=names.count().sort_values(by=['playerID'],ascending=False)
namesList.head(25)
```

```python
data.sort_values(by=['lastName'],ascending=True)
```

```python
players=data[["playerID", "year","GP","G","A"]].groupby(["playerID", "year"]).sum()
stat=players.groupby("playerID").cumsum()
stat=pd.DataFrame(stat)
stat.head(50)
stat.to_csv("SumofGoals.csv")
```
